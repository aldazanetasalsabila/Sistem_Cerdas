```python
PNoDeliveryNo = 4/9
PYesDeliveryNo = 5/9

PNoDeliveryYes = 10/41
PYesDeliveryYes = 31/41

PNoDelivery = 14/50
PYesDelivery = 36/50
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = NoBuy = PNo
# P(B) = PYesDiscount,PYesDelivery,PHoliday
# P(B|A)1 = (PHoliday|NoBuy) = PHolidayNo
# P(B|A)2 = (PYesDiscount|NoBuy) = PYesDiscountNo
# P(B|A)3 = (PYesDelivery|NoBuy) = PYesDeliveryNo
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

#ditanyakan P(A|B) ?
#Rumus
# P(A|B) = P(B|A)*P(A)/P(B)
```

```python
PNoBuy1 = ((PHolidayNo*PYesDiscountNo*PYesDeliveryNo)*PNo)/(PHoliday*PYesDiscount*PYesDelivery)
print(PNoBuy1)
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = Buy = PYes
# P(B) = PYesDiscount,PYesDelivery,PHoliday
# P(B|A)1 = (PHoliday|Buy) = PHolidayYes
# P(B|A)2 = (PYesDiscount|Buy) = PYesDiscountYes
# P(B|A)3 = (PYesDelivery|Buy) = PYesDeliveryYes
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

#ditanyakan P(A|B) ?
#Rumus
# P(A|B) = P(B|A)*P(A)/P(B)
```

```python
PBuy2 = ((PHolidayYes*PYesDiscountYes*PYesDeliveryYes)*PYes)/(PHoliday*PYesDiscount*PYesDelivery)
print(PBuy2)
```

```python
#Probability of No Buy = 0.12989982125784594
#Probability of Buy = 0.8440667832398943
```

```python
# P(Buy | Day=Weekday, FreeDelivery=Yes, Discount=Yes)

PBuy_a = ((PWeekdayYes * PYesDeliveryYes * PYesDiscountYes) * PYes) / (PWeekday * PYesDelivery * PYesDiscount)
PNoBuy_a = ((PWeekdayNo * PYesDeliveryNo * PYesDiscountNo) * PNo) / (PWeekday * PYesDelivery * PYesDiscount)

PSum_a = PBuy_a + PNoBuy_a

PBuy_a_final = (PBuy_a / PSum_a) * 100
PNoBuy_a_final = (PNoBuy_a / PSum_a) * 100

print("a. P(Buy | Weekday, Yes, Yes)     =", PBuy_a_final)
print("c. P(Not Buy | Weekday, Yes, Yes) =", PNoBuy_a_final)
```

```python
# P(Buy | Day=Weekday, FreeDelivery=No, Discount=No)

PBuy_b = ((PWeekdayYes * PNoDeliveryYes * PNoDiscountYes) * PYes) / (PWeekday * PNoDelivery * PNoDiscount)
PNoBuy_b = ((PWeekdayNo * PNoDeliveryNo * PNoDiscountNo) * PNo) / (PWeekday * PNoDelivery * PNoDiscount)

PSum_b = PBuy_b + PNoBuy_b

PBuy_b_final = (PBuy_b / PSum_b) * 100
PNoBuy_b_final = (PNoBuy_b / PSum_b) * 100

print("b. P(Buy | Weekday, No, No)     =", PBuy_b_final)
print("d. P(Not Buy | Weekday, No, No) =", PNoBuy_b_final)
```

```python
# P(Buy | Day=Weekend, FreeDelivery=Yes, Discount=Yes)

PBuy_e = ((PWeekendYes * PYesDeliveryYes * PYesDiscountYes) * PYes) / (PWeekend * PYesDelivery * PYesDiscount)
PNoBuy_e = ((PWeekendNo * PYesDeliveryNo * PYesDiscountNo) * PNo) / (PWeekend * PYesDelivery * PYesDiscount)

PSum_e = PBuy_e + PNoBuy_e

PBuy_e_final = (PBuy_e / PSum_e) * 100
PNoBuy_e_final = (PNoBuy_e / PSum_e) * 100

print("e. P(Buy | Weekend, Yes, Yes)     =", PBuy_e_final)
print("g. P(Not Buy | Weekend, Yes, Yes) =", PNoBuy_e_final)
```

```python
# P(Buy | Day=Weekend, FreeDelivery=No, Discount=No)

PBuy_f = ((PWeekendYes * PNoDeliveryYes * PNoDiscountYes) * PYes) / (PWeekend * PNoDelivery * PNoDiscount)
PNoBuy_f = ((PWeekendNo * PNoDeliveryNo * PNoDiscountNo) * PNo) / (PWeekend * PNoDelivery * PNoDiscount)

PSum_f = PBuy_f + PNoBuy_f

PBuy_f_final = (PBuy_f / PSum_f) * 100
PNoBuy_f_final = (PNoBuy_f / PSum_f) * 100

print("f. P(Buy | Weekend, No, No)     =", PBuy_f_final)
print("h. P(Not Buy | Weekend, No, No) =", PNoBuy_f_final)
```
